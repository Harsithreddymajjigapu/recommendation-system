```python
df['Outlet_Age'] = 2013 - df['Outlet_Establishment_Year']
```

```python
cols = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Outlet_Age', 'Outlet_Age']

for col in cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    outlierscount=df[(df[col]>upper)|(df[col]<lower)]
    print(len(outlierscount))
```

```python
df['Outlet_Size'] = df['Outlet_Size'].map({'Small':0, 'Medium':1, 'High':2})
df['Outlet_Location_Type'] = df['Outlet_Location_Type'].map({'Tier 1':0, 'Tier 2':1, 'Tier 3':2})
df['Item_Fat_Content'] = df['Item_Fat_Content'].map({'Low Fat':0, 'Regular':1, 'Non-Edible':2})
```
